The data behind the public websites of the GRAPHIKPORTAL have been recently made available for programmatic approach as a set of GitLab repositories available from here: https://gitlab.uni-marburg.de/fotomarburg/graphikportal/lab.

Individual data entries ara availble within the `LIDO` subdirectories of each repository in the form of **LIDO XML records**. – a metadata schema used for describing museum and cultural heritage objects. It stands for “Lightweight Information Describing Objects” and is a standard for exchanging information between collections and portals (like Europeana, museum databases, or aggregators).

```python
from sickle import Sickle
from sickle.models import Record
import requests
from pathlib import Path
```

```python
# Set up
base_url = "https://sammlung.staedelmuseum.de/api/oai"
output_dir = Path("lido_records")
output_dir.mkdir(exist_ok=True)

# Initialize OAI-PMH client
sickle = Sickle(base_url)

records = {}
# Step 1: List identifiers
identifiers = sickle.ListIdentifiers(metadataPrefix="lido")
```

```python
# Step 2: Fetch and store
for i, header in enumerate(identifiers):
    identifier = header.identifier
    object_id = identifier.split(":")[-1]
    print(f"[{i}] Fetching: {identifier}")

    try:
        record = sickle.GetRecord(metadataPrefix="lido", identifier=identifier)
        records[object_id] = record

        # Optional: Save raw XML to file
        with open(output_dir / f"{object_id}.xml", "w", encoding="utf-8") as f:
            f.write(record.raw)

    except Exception as e:
        print(f"⚠️ Skipping {identifier}: {e}")
```

```python

# Step 2: Fetch each record
for i, header in enumerate(identifiers):
    identifier = header.identifier
    print(f"[{i}] Fetching: {identifier}")

    # Fetch LIDO record
    record = sickle.GetRecord(metadataPrefix="lido", identifier=identifier)

    # Save to file
    object_id = identifier.split(":")[-1]
    records[object_id] = record
```
